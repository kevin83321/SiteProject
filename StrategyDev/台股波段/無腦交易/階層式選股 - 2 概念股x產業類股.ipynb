# 從資料庫取得資料

```python
client = getclient()
schema = client['admin']
```

## 從資料庫取得概念股族群

```python
group_table = schema['ConceptGroups']
update_date = sorted(group_table.distinct("UpdateDate"))[-1]
groups_data = pd.DataFrame(list(group_table.find({"UpdateDate":{"$eq":update_date}})))
del groups_data['_id']
```

```python
groups_data
```

## 從資料庫取得產業類股

```python
info_table = schema['TWSE.StockList']
update_date_info = sorted(info_table.distinct("UpdateDate"))[-1]
info_data = pd.DataFrame(list(info_table.find({"UpdateDate":{"$eq":update_date_info}, 
                                               'AssetType':{"$in":["股票", "台灣存託憑證(TDR)",
                                                                   "台灣存託憑證", "普通股"]}})))
del info_data['_id']
```

```python
info_data
```

## 從資料庫取得類股指數

```python
index_table = schema['TWSE.HistoricalPrice.Index.Interday']
last_date = sorted(index_table.distinct("Date"))[-1]
index_data = pd.DataFrame(index_table.find({"Date":{"$eq":last_date}}))
del index_data['_id']
```
